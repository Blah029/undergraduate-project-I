```python
# Input and output folder paths
input_folder = '/content/gdrive/MyDrive/4th Year/7th Semester/EE405 Undergraduate Project I/G-08-2023/Shared Items/Datasets/Labelme Dataset 01/masks'  # Specify the path to the folder containing RGB images
output_folder = '/content/gdrive/MyDrive/4th Year/7th Semester/EE405 Undergraduate Project I/G-08-2023/Shared Items/Datasets/Labelme Dataset 01/masks_processed'  # Specify the path to the folder for saving red channel images

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through the files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.png') or filename.endswith('.jpg'):  # Process only PNG and JPG images
        # Read the image
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path, cv2.IMREAD_COLOR)

        # Extract the red channel
        red_channel = image[:, :, 2]  # Access the red channel

        # Save the red channel image in the output folder
        output_path = os.path.join(output_folder,filename[:-4]+"_processed"+filename[-4:])
        cv2.imwrite(output_path, red_channel)
```
